# Check for the unique values


```python
# Check for how many unique values are in a column
unique_values = df.nunique()
unique_values.sort_values(ascending = False)
```

```python
## get the unique values from the columns for look ups
## option 1: include columns with more than 1 unique value, but if you include columns like party_id, you will end up with the really long table, so you could set up a range ''if 50 > len(unique_values) > 1:''
def unique_values_table(df):
    unique_values_dict = {}
    for col in df.columns:
        unique_values = df[col].unique()
        if 50 > len(unique_values) > 1:  ###### change the range here
            unique_values_dict[col] = unique_values

    if unique_values_dict:
        # Find the maximum number of unique values among all columns
        max_unique_values = max(len(values) for values in unique_values_dict.values())

        # Create a dictionary to store the data for the DataFrame
        data = {}
        for col, values in unique_values_dict.items():
            data[col] = values.tolist() + [''] * (max_unique_values - len(values))

        # Create the DataFrame
        return pd.DataFrame(data)
    else:
        return pd.DataFrame()

# Get unique values table as DataFrame
unique_values_df = unique_values_table(df)
```

```python
## option 2: only select a few colums you are interested in
def unique_values_table(df, columns=None):
    if columns is None:
        columns = df.columns

    unique_values_dict = {}
    for col in columns:
        unique_values = df[col].unique()
        if len(unique_values) > 1:
            unique_values_dict[col] = unique_values

    if unique_values_dict:
        # Find the maximum number of unique values among all columns
        max_unique_values = max(len(values) for values in unique_values_dict.values())

        # Create a dictionary to store the data for the DataFrame
        data = {}
        for col, values in unique_values_dict.items():
            data[col] = values.tolist() + [''] * (max_unique_values - len(values))

        # Create the DataFrame
        return pd.DataFrame(data)
    else:
        return pd.DataFrame()
```

```python
## run the function
unique_values_df = unique_values_table(df, columns= ['call_reason_1', 'call_reason_2']) ### change here for column names
unique_values_df
```
